```python
import pandas as pd
from coinmarketcap import Market
import numpy as np
```

```python
coinmarketcap = Market()
market_summary = coinmarketcap.ticker(limit=100)
cmc_df = pd.DataFrame(market_summary)
cmc_df
  
```

```python
# LEt pull out the neccesssary columns create a new Dataframe then, we will sort and filter for our info 
slim_cmc_df = cmc_df[["percent_change_24h","id","24h_volume_usd","rank", "price_usd", "price_btc"]].sort_values("percent_change_24h", ascending=True)
slim_cmc_df.head()
```

```python
slim_cmc_df = slim_cmc_df.head(n=10)
trending = ["1st","2nd","3rd","4th","5th","6th","7th","8th","9th","10th"]
slim_cmc_df.index=trending 
slim_cmc_df
#slim_cmc_df_reindexed.head(n=10)
#slim_cmc_df_reindexed
```
